```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("log_eth1.csv")
```

```python
plt.show(df.plot(x='t',y=['spot_price','target_price']))
```

```python
plt.show(df.plot(x='t',y='act_ETH'))
```

```python
df['avg_10'] = df[['spot_price']].rolling(10).mean()
df['avg_100'] = df[['spot_price']].rolling(100).mean()
df['avg_1000'] = df[['spot_price']].rolling(1000).mean()
df['avg_3000'] = df[['spot_price']].rolling(3000).mean()
```

```python
plt.show(df.plot(x='t',y=['spot_price','target_price','avg_100']))
```

```python
len(df[['spot_price']].iloc[1:])
```

```python
df[['spot_price']].rolling(2).mean().count()
```

```python
min_prd = 1
max_prd = 3000
errors = []
for n in range(min_prd,max_prd+1):
    roll = df[['spot_price']].rolling(n).mean()
    res = (roll - df[['spot_price']])**2
    e = float(res.iloc[max_prd-1:].sum())
    errors.append({'n': n, 'error': e})
    print(float(res.iloc[max_prd-1:].sum()))
edf = pd.DataFrame(errors)
```

```python
edf.head()
```

```python
plt.show(edf.plot(x='n',y='error'))
```
